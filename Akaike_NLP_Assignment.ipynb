**REMOVING ROMAN LETTER(NUMBER)**

```python
def clean(text):
    pattern = r"\b(?=[MDCLXVIΙ])M{0,4}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})([IΙ]X|[IΙ]V|V?[IΙ]{0,3})\b\.?"
    return re.sub(pattern, '', text)
```

```python
twt=[]
for i in df1["tweet_text"]:
  twt.append(clean(i))
```

### **COUNTVECTORIZER**

Import the libraries we need
give you the bag of words model

Step 1. Get the data

Step 2. Design the Vocabulary. The default token pattern removes tokens of a single character. That's why we don't have the "I" and "s" tokens in the output.

Step 3. Create the Bag-of-Words Model. Fit - design the vocbulary and transform will convert the text into numbers based on the presence of the word

Step 4. Show the Bag-of-Words Model as a pandas DataFrame

```python
from sklearn.feature_extraction.text import CountVectorizer 
import pandas as pd
fn=[]
bw=[]
for i in words_text:
  documents=[i]

  count_vectorizer = CountVectorizer(binary = False) 

  bag_of_words = count_vectorizer.fit_transform(documents) 
  bw.extend(*(bag_of_words.toarray()))

  feature_names = count_vectorizer.get_feature_names()
  fn.extend(feature_names)
```
